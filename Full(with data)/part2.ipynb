```python
def tier_store_performance(df, start_month, end_month, thresholds):
    # Filter data within the specified time range
    mask = (df['Month'] >= start_month) & (df['Month'] <= end_month)
    filtered_df = df.loc[mask]
    
    # Calculate the average growth rate for each store
    store_performance = filtered_df.groupby('Product Line')['Growth'].mean()
    
    # Assign tiers based on the average growth rate
    store_tiers = {}
    for store, growth in store_performance.items():
        tier = 'Tier 1'
        for i, threshold in enumerate(thresholds, start=1):
            if growth < threshold:
                tier = f'Tier {i+1}'
                break
        store_tiers[store] = tier
    
    return store_tiers
```

```python
import pandas as pd
from geopy.geocoders import OpenCage

def add_coordinates_column(df, api_key):
    # Initialize the OpenCage Geocoder
    geolocator = OpenCage(api_key=api_key)
    
    # Function to retrieve coordinates based on address
    def get_coordinates(row):
        address = f"{row['Parent Store Address']}, {row['Parent Store City Name']}, {row['Parent Store State']}, {row['Parent Store Postal Cd']}, {row['Parent Store Country']}"
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None
    
    # Apply the get_coordinates function to each row and create a new column
    df['Coordinates'] = df.apply(get_coordinates, axis=1)
    
    return df

# Usage
api_key = '045342dbdc1141289b8a04dd9652277d' 
df_add = add_coordinates_column(df_add, api_key)
```

```python
from geopy.distance import geodesic

def assign_warehouse(df_add):
    warehouses = {
        'Olive Branch, MS': (34.9626, -89.8284),  # 8640 Nail Rd, Olive Branch, MS 38654
        'Henderson, CO': (39.9138, -104.8698)     # 9801 Dallas St, Henderson, CO 80640
    }
    

    
    def get_closest_warehouse(store_coords):
        min_distance = float('inf')
        closest_warehouse = None
        for warehouse, warehouse_coords in warehouses.items():
            distance = geodesic(store_coords, warehouse_coords).miles
            if distance < min_distance:
                min_distance = distance
                closest_warehouse = warehouse
        return closest_warehouse
    
    df_add['Assigned Warehouse'] = df_add['Coordinates'].apply(get_closest_warehouse)
    
    return df_add
```
