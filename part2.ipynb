```python
import csv
from collections import defaultdict
from datetime import datetime

def calculate_yoy_growth(data):
    #Structure: product_lines[product_line][month][parent_store] = sales_value
    product_lines = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
    
    for row in data:
        parent_store = row[0]
        product_sku = row[1]
        product_line = row[-1]

        for i in range(2, len(row) - 1):
            if row[i]:
                month = datetime.strptime(header[i], '%B %Y').strftime('%Y-%m')
                value = int(float(row[i].replace(',', '')))  # Remove commas (some values are like '1,295') and convert to float, then to int
                product_lines[product_line][month][int(parent_store)] += value
                
    # Structure: yoy_growth[product_line][month] = growth_percentage
    yoy_growth = defaultdict(lambda: defaultdict(dict))

    for product_line, monthly_data in product_lines.items():
        months = sorted(monthly_data.keys())

        for i in range(12, len(months)):
            current_month = months[i]
            prev_year_month = months[i - 12]
            
            # Calculate the total sales for the current month and the same month in the previous year
            current_total = sum(monthly_data[current_month].values())
            prev_year_total = sum(monthly_data[prev_year_month].values())

            if prev_year_total != 0:
                growth = (current_total - prev_year_total) / prev_year_total * 100
                yoy_growth[product_line][current_month] = growth

    return yoy_growth

with open('combined_df.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    data = list(reader)

# Calculate year-over-year growth
yoy_growth = calculate_yoy_growth(data)

# Print the year-over-year growth for each product line and month
for product_line, growth_data in yoy_growth.items():
    print(f"Product Line: {product_line}")
    for month, growth in growth_data.items():
        print(f"  {month}: {growth:.2f}%")
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'yoy_growth.csv' is already in your working directory
file_path_per = 'yoy_growth.csv'
df_per = pd.read_csv(file_path_per)

# Convert the 'Month' column to datetime
df_per['Month'] = pd.to_datetime(df_per['Month'])

# Calculate the average growth for each product line
average_growth = df_per.groupby('Product Line')['Growth'].mean().sort_values(ascending=False)

# Selecting a few product lines for visualization
# Top 2 performers
top_performers = average_growth.head(2).index.tolist()

# Median performer
median_index = average_growth.shape[0] // 2
median_performer = average_growth.iloc[[median_index]].index.tolist()

# Bottom performer
bottom_performer = average_growth.tail(1).index.tolist()

# Compile the list of product lines to be visualized
selected_product_lines = top_performers + median_performer + bottom_performer

# Filter the dataframe for the selected product lines
df_selected = df_per[df_per['Product Line'].isin(selected_product_lines)]

# Create a Line Chart for selected product lines
fig, ax = plt.subplots(figsize=(12, 6))

for product_line in selected_product_lines:
    df_product = df_selected[df_selected['Product Line'] == product_line]
    ax.plot(df_product['Month'], df_product['Growth'], label=product_line)

ax.set_title('Year-over-Year Sales Growth for Selected Product Lines')
ax.set_xlabel('Month')
ax.set_ylabel('Growth Percentage (%)')
ax.legend(loc='upper left')  # Move the legend to the upper left for better visibility
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the layout

plt.show()
```

```python
import pandas as pd
from geopy.geocoders import OpenCage

def add_coordinates_column(df, api_key):
    # Initialize the OpenCage Geocoder
    geolocator = OpenCage(api_key=api_key)
    
    # Function to retrieve coordinates based on address
    def get_coordinates(row):
        address = f"{row['Parent Store Address']}, {row['Parent Store City Name']}, {row['Parent Store State']}, {row['Parent Store Postal Cd']}, {row['Parent Store Country']}"
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None
    
    # Apply the get_coordinates function to each row and create a new column
    df['Coordinates'] = df.apply(get_coordinates, axis=1)
    
    return df

# Usage
api_key = '045342dbdc1141289b8a04dd9652277d' 
df_add = add_coordinates_column(df_add, api_key)
```

```python
from geopy.distance import geodesic

def assign_warehouse(df_add):
    warehouses = {
        'Olive Branch, MS': (34.9626, -89.8284),  # 8640 Nail Rd, Olive Branch, MS 38654
        'Henderson, CO': (39.9138, -104.8698)     # 9801 Dallas St, Henderson, CO 80640
    }
    

    
    def get_closest_warehouse(store_coords):
        min_distance = float('inf')
        closest_warehouse = None
        for warehouse, warehouse_coords in warehouses.items():
            distance = geodesic(store_coords, warehouse_coords).miles
            if distance < min_distance:
                min_distance = distance
                closest_warehouse = warehouse
        return closest_warehouse
    
    df_add['Assigned Warehouse'] = df_add['Coordinates'].apply(get_closest_warehouse)
    
    return df_add
```
